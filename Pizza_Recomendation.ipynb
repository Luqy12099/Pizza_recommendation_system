```python
raw_df = pd.read_excel('/content/pizza/Pizza Flavors Sheet.xlsx')
raw_df.head()
```

# Preprocessing

```python
LISTNAMEOFPIZZA = raw_df['Name of Pizza'].to_list()
index = LISTNAMEOFPIZZA.index(np.nan)
df = raw_df.iloc[:index,:]
df.head()
```

# Pearson correlation

```python
def pearson(df, cust):
    hitungPearson = []
    totalSim = []
    a = 0
    x = raw_df.iloc[cust, 1:].to_list()#ganti cust
    for NameOfPizza in range(df.shape[0]):
        Pizza = df.iloc[NameOfPizza, 1:].to_list()
        sim = scipy.stats.pearsonr(x, Pizza)
        sim = sim[0]
        totalSim.append(sim)
    return totalSim
```

```python
def sortBestRecommendation(totalSim):
    sim = pd.DataFrame(totalSim, columns=['sim'])
    sim.sort_values(by=['sim'], inplace=True, ascending=False)
    sim = sim.reset_index()
    sim = sim.iloc[1:4,0].to_list()
    return sim
```

# run proses input
